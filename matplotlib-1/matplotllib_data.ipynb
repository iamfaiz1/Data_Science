# create a live chart to find sale trends

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('matplotlib_data.csv')
df.head()
df.tail()
df.info()
df.describe()

# line plot example
df['Sales'].plot(kind='line', title = 'Line Plot of Sales over')
df['Profit'].plot(kind='line', title = 'Line Plot of Profit over')
plt.xlabel('Index')
plt.ylabel('Sales')
plt.show()
```

**Q2: Average sales per category**

```python
# group by example
df.groupby('Category')['Sales'].mean().plot(
    kind='bar', 
    color = 'grey',
    title='Average Sales by Category'
    )
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.legend(['Sales'])
plt.show()
```

```python
# group by example
df.groupby('Category')['Profit'].mean().plot(
    kind='bar', 
    color = 'grey',
    title='Average Profit by Category'
    )
plt.xlabel('Category')
plt.ylabel('Average Profit')
plt.legend(['Profit'])
plt.show()
```

**pie chart example for profit by category**

```python
category_counts = df['Category'].value_counts()
category_counts.plot(
    ylabel='',
    kind = 'pie',
    autopct = '%1.2f%%',    #shows the percentage digits
    startangle = 90,
    title = 'Category Distribution',
    colors = ['lightblue', 'lightgreen', 'lightcoral', 'gold' ]
)

# plt.ylabel('')
plt.show()
```

# create histogram for sales figure

```python
# histogram example
df['Sales'].plot(
    kind = 'hist',
    bins = 20,
    color = 'lightpink',
    title= 'histogram of Sales',
    alpha = 0.7,
)
plt.xlabel('Sales')
plt.show()
```

```python
# subplots example

fig, axs = plt.subplots(1, 2, figsize=(10, 4))   

# fig : the overall figure object (the canvas that holds all the subplots)
# axs : an array of axes objects (one for each subplot)

axs[0].plot(
    df['Sales'], 
    color='blue',
    linestyle='--'
    )

axs[1].plot(
    df['Profit'],
    color ='green',
    linestyle=':'
)

plt.show()
```

```python
# example 2
fig , axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(
    df['Sales'], 
    color='blue',
    linestyle='--'
)
axs[0,0].set_title('Sales')


axs[0, 1].plot(
    df['Sales'], 
    color ='green',
    linestyle=':'
)
axs[0,1].set_title('Sales')


axs[1, 0].plot(
    df['Sales'], 
    color ='red',
    linestyle='-.'
)
axs[1,0].set_title('Sales')
```

```python
# different graphs in all 4

fig, axs = plt.subplots(2, 2, figsize=(10, 10))  # Create a 2x2 grid of subplots)

# line plot
axs[0, 0].plot(
    df['Sales'], 
    color = 'blue',
    linestyle = '--',
)
axs[0,0].set_title('Line Plot of Sales')


# bar plot
axs[0, 1].bar(
df['Category'],               # x-axis: index of the dataframe
df['Profit'],                  # y-axis: profit values (height)
color = 'lightgreen',
alpha = 0.7,
)
axs[0, 1].set_title('Bar Plot of Profit')
axs[0, 1].set_ylabel('Profit')
axs[0, 1].set_xlabel('Category')


# pie-chart
category_counts = df['Category'].value_counts()
axs[1, 0].pie(
    category_counts, 
    labels=category_counts.index, 
    autopct='%1.1f%%', 
    colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
axs[1, 0].set_title('Category Distribution')

# Histogram
axs[1, 1].hist(
    df['Sales'], 
    bins=20, 
    color='purple')
axs[1, 1].set_title('Sales Histogram')
```
